```python
tr, Utr, ts, Uts = rc.train_test_orbit("dscc", duration=50, dt=0.001)
```

```python
rcomp_default = rc.ResComp()
rcomp_default.train(tr, Utr)
```

```python
r0 = rcomp_default.r0
pre = rcomp_default.predict(ts, r0=r0)
fig = rc.plot3d(pre)
```

```python
hyper_parameters = rc.SYSTEMS["dscc"]["rcomp_params"]
rcomp = rc.ResComp(**hyper_parameters)
```

```python
rcomp.train(tr, Utr)
r0 = rcomp.r0
pre = rcomp.predict(ts, r0=r0)
fig = rc.plot3d(pre)
```
